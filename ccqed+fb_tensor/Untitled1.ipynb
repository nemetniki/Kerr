```python
import numpy as np

indS = 4
indB = 2
B = np.ones(5)
F = np.ones(5)
S = np.ones(9)
B2 = np.zeros(5)
B2[indB] = 1
F2 = np.zeros(5)
F2[indB] = 1
S2 = np.zeros(9)
S2[indS] = 1
print(S)
state1 = np.einsum("i,j,k,l,m,n->ijklmn",F,S,B,F,S,B)
state2 = np.einsum("i,j,k,l,m,n->ijklmn",F2,S2,B2,F2,S2,B2)
dim_tS = S.shape[0]
M=0
es=np.einsum
```

```python
def a(state,which,N):
    new_state = np.zeros(state.shape,complex)
    if dim_tS>2*N:
        n=np.linspace(N,int((dim_tS)/2),dim_tS-2*N).astype(np.int64)
        nprod = 1.
        i = 0
        while i<N:
            nprod *= np.sqrt(n-i)
            i+=1

        ich  = (-1)**(which+1)*(2*int(np.any(M))*(2-which)+which)#which=1->1/3, which=2->-2/-2
        iend = state.shape[ich]
        context = ["j,ijklmn->ijklmn","m,ijklmn->ijklmn","J,gIeJKhLfMN->gIeJKhLfMN","M,gIeJKhLfMN->gIeJKhLfMN"]

        idx=[slice(None)]*state.ndim
        idx[ich]=slice(None,iend-2*N)

        new_state[tuple(idx)] = es(context[which-1+int(np.any(M))*2],nprod,np.take(state,np.arange(2*N,iend),axis=ich))

    return new_state

print(state1[0,:,0,0,0,0])
print(a(state1,1,1)[0,:,0,0,0,0]**2)
print(state1[0,0,0,0,:,0])
print(a(state1,2,1)[0,0,0,0,:,0]**2)
print(state1[0,:,0,0,0,0])
print(a(state1,1,2)[0,:,0,0,0,0]**2)
print(state1[0,0,0,0,:,0])
print(a(state1,2,2)[0,0,0,0,:,0]**2)

print("\n")
print(state2[indB,:,indB,indB,indS,indB])
print(a(state2,1,1)[indB,:,indB,indB,indS,indB]**2)
print(state2[indB,indS,indB,indB,:,indB])
print(a(state2,2,1)[indB,indS,indB,indB,:,indB]**2)
print(state2[indB,:,indB,indB,indS,indB])
print(a(state2,1,2)[indB,:,indB,indB,indS,indB]**2)
print(state2[indB,indS,indB,indB,:,indB])
print(a(state2,2,2)[indB,indS,indB,indB,:,indB]**2)
```

```python
def sm(state,which):
    new_state = np.zeros(state.shape,complex)
    ich  = (-1)**(which+1)*(2*int(np.any(M))*(2-which)+which)#which=1->1/3, which=2->-2/-2
    iend = state.shape[ich]

    idx=[slice(None)]*state.ndim
    idx[ich]=slice(0,dim_tS-1,2)

    new_state[tuple(idx)] = np.take(state,np.arange(1,dim_tS,2),axis=ich)
    return new_state
print(state1[0,:,0,0,0,0])
print(sm(state1,1)[0,:,0,0,0,0]**2)
print(state1[0,0,0,0,:,0])
print(sm(state1,2)[0,0,0,0,:,0]**2)

print("\n")
print(state2[indB,:,indB,indB,indS,indB])
print(sm(state2,1)[indB,:,indB,indB,indS,indB]**2)
print(state2[indB,indS,indB,indB,:,indB])
print(sm(state2,2)[indB,indS,indB,indB,:,indB]**2)
```

```python
def sp(state,which):
    new_state = np.zeros(state.shape,complex)
    ich  = (-1)**(which+1)*(2*int(np.any(M))*(2-which)+which)#which=1->1/3, which=2->-2/-2
    iend = state.shape[ich]

    idx=[slice(None)]*state.ndim
    idx[ich]=slice(1,dim_tS,2)

    new_state[tuple(idx)] = np.take(state,np.arange(0,dim_tS-1,2),axis=ich)
    return new_state

print(state1[0,:,0,0,0,0])
print(sp(state1,1)[0,:,0,0,0,0]**2)
print(state1[0,0,0,0,:,0])
print(sp(state1,2)[0,0,0,0,:,0]**2)

print("\n")
print(state2[indB,:,indB,indB,indS,indB])
print(sp(state2,1)[indB,:,indB,indB,indS,indB]**2)
print(state2[indB,indS,indB,indB,:,indB])
print(sp(state2,2)[indB,indS,indB,indB,:,indB]**2)
```

```python
def JC(state,which):
    new_tS_g = np.zeros(state.shape,complex)
    new_tS_Ome = np.zeros(state.shape,complex)
    n = np.linspace(1,int((dim_tS)/2),dim_tS-2).astype(np.int64)
    n = np.sqrt(n[0:dim_tS-1:2])

    ich  = (-1)**(which+1)*(2*int(np.any(M))*(2-which)+which)#which=1->1/3, which=2->-2/-2
    iend = state.shape[ich]

    if g[which-1] != 0:
        context = ["j,ijklmn->ijklmn","m,ijklmn->ijklmn","J,gIeJKhLfMN->gIeJKhLfMN","M,gIeJKhLfMN->gIeJKhLfMN"]

        idx1=[slice(None)]*state.ndim
        idx2=[slice(None)]*state.ndim
        idx1[ich]=slice(1,dim_tS-1,2)
        idx2[ich]=slice(2,dim_tS,2)

        new_tS_g[tuple(idx1)] = es(context[which-1+int(np.any(M))*2],n*g[which-1],np.take(state,np.arange(2,dim_tS,2),axis=ich))
        new_tS_g[tuple(idx2)] = es(context[which-1+int(np.any(M))*2],n*g[which-1],np.take(state,np.arange(1,dim_tS-1,2),axis=ich))

    if Ome[which-1] != 0:
        idx1=[slice(None)]*state.ndim
        idx2=[slice(None)]*state.ndim
        idx1[ich]=slice(0,dim_tS-2,2)
        idx2[ich]=slice(1,dim_tS-1,2)

        new_tS_Ome[tuple(idx1)] = np.take(state,np.arange(1,dim_tS-1,2),axis=ich)
        new_tS_Ome[tuple(idx2)] = np.take(state,np.arange(0,dim_tS-2,2),axis=ich)
    return new_tS_g+new_tS_Ome

g=np.array([1,1])
Ome = np.array([0,0])

print(state1[0,:,0,0,0,0])
print(JC(state1,1)[0,:,0,0,0,0]**2)
print(state1[0,0,0,0,:,0])
print(JC(state1,2)[0,0,0,0,:,0]**2)

print("\n")
print(state2[indB,:,indB,indB,indS,indB])
print(JC(state2,1)[indB,:,indB,indB,indS,indB]**2)
print(state2[indB,indS,indB,indB,:,indB])
print(JC(state2,2)[indB,indS,indB,indB,:,indB]**2)
```

```python
def nc(state,which,const):
    n=np.linspace(0,int((dim_tS)/2),dim_tS).astype(np.int64)
    new_state = np.zeros(state.shape,complex)
    context = ["j,ijklmn->ijklmn","m,ijklmn->ijklmn","J,gIeJKhLfMN->gIeJKhLfMN","M,gIeJKhLfMN->gIeJKhLfMN"]

    new_state = es(context[which-1+int(np.any(M))*2],n+const,state)
    return new_state

print(state1[0,:,0,0,0,0])
print(nc(state1,1,0)[0,:,0,0,0,0])
print(state1[0,0,0,0,:,0])
print(nc(state1,2,0)[0,0,0,0,:,0])

print("\n")
print(state2[indB,:,indB,indB,indS,indB])
print(nc(state2,1,0)[indB,:,indB,indB,indS,indB])
print(state2[indB,indS,indB,indB,:,indB])
print(nc(state2,2,0)[indB,indS,indB,indB,:,indB])
```

```python
def nE(state,which,const):
    new_state_tF = np.zeros(state.shape,complex)
    new_state_tB = np.zeros(state.shape,complex)
    new_state_tFB1 = np.zeros(state.shape,complex)
    new_state_tFB2 = np.zeros(state.shape,complex)
    n = np.arange(0,state.shape[-1])
    if gamma_F[which-1]>0:
        context = ["i,ijklmn->ijklmn","l,ijklmn->ijklmn","I,gIeJKhLfMN->gIeJKhLfMN","L,gIeJKhLfMN->gIeJKhLfMN"]

        new_state_tF = dt*gamma_F[which-1]*es(context[which-1+2*int(np.any(M))],n+const,state)

    if gamma_B[which-1]>0:
        context = ["k,ijklmn->ijklmn","n,ijklmn->ijklmn","K,gIeJKhLfMN->gIeJKhLfMN","N,gIeJKhLfMN->gIeJKhLfMN"]

        new_state_tB = dt*gamma_B[which-1]*es(context[which-1+2*int(np.any(M))],n+const,state)

    if gamma_B[which-1]>0 and gamma_F[which-1]>0:

        if M<L:
            phip=0.
        else:
            phip=phi

        context = ["i,k,ijklmn->ijklmn","l,n,ijklmn->ijklmn","I,K,gIeJKhLfMN->gIeJKhLfMN","L,N,gIeJKhLfMN->gIeJKhLfMN"]
        ich1 = (-1)**(which+1)*(int(np.any(M))*2*(2-which)+(3-which))#which=1->2/4, which=2->-1/-1
        ich2 = int(np.any(M))*(2-which)+(int(np.any(M))+3)*(1-which)#which=1->0/1, which=2->-3/-4
        iend1 = state.shape[ich1]
        iend2 = state.shape[ich2]
        n = np.sqrt(n[1:])

        idx_n1=[slice(None)]*state.ndim
        idx_n1[ich1]=slice(None,iend1-1)
        idx_n1[ich2]=slice(1,iend2)
        idx_n2=[slice(None)]*state.ndim
        idx_n2[ich1]=slice(1,iend1)
        idx_n2[ich2]=slice(None,iend2-1)
        idx_o1=[slice(None)]*state.ndim
        idx_o1[ich1]=slice(1,iend1)
        idx_o1[ich2]=slice(None,iend2-1)
        idx_o2=[slice(None)]*state.ndim
        idx_o2[ich1]=slice(None,iend1-1)
        idx_o2[ich2]=slice(1,iend2)

        new_state_tFB1[tuple(idx_n1)] = (np.sqrt(gamma_F[which-1]*gamma_B[which-1])*dt*np.exp(1j*phip)*
                        contract(context[which-1+int(np.any(M))*2],n,n,state[tuple(idx_o1)]))
        new_state_tFB2[tuple(idx_n2)] = (np.sqrt(gamma_F[which-1]*gamma_B[which-1])*dt*np.exp(-1j*phip)*
                        contract(context[which-1+int(np.any(M))*2],n,n,state[tuple(idx_o2)]))
        

    return new_state_tF+new_state_tB+new_state_tFB1+new_state_tFB2

print(state1[:,0,0,0,0,0])
print(nE(state1,1,0)[:,0,1,1,0,1])
print(state1[0,0,:,0,0,0])
print(nE(state1,1,0)[1,0,:,1,0,1])
print(state1[0,0,0,:,0,0])
print(nE(state1,2,0)[1,0,1,:,0,1])
print(state1[0,0,0,0,0,:])
print(nE(state1,2,0)[1,0,1,1,0,:])

print("\n")
print(state2[:,indS,indB,indB,indS,indB])
print(nE(state2,1,0)[:,indS,indB-1,indB,indS,indB])
print(nE(state2,1,0)[indB,indS,indB+1,:,indS,indB])
print(state2[indB,indS,:,indB,indS,indB])
print(nE(state2,1,1)[indB,indS,:,indB,indS,indB])
print(state2[indB,indS,indB,:,indS,indB])
print(nE(state2,2,2)[indB,indS,indB,:,indS,indB])
print(state2[indB,indS,indB,indB,indS,:])
print(nE(state2,2,0)[indB,indS,indB,indB,indS,:])
```
