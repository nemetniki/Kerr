```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

num = 2
diss=True

Data = np.zeros((1501,8))
Data[:,:] = np.genfromtxt('../Data/CCQED+fb_%d.txt'%num)
Data2 = np.zeros((1501,6))
Data2[:,:] = np.genfromtxt('../Data/CCQED_MC%d.txt'%num)

if diss:
    Data3 = np.zeros((1501,5))
    Data3[:,:] = np.genfromtxt('../Data/CCQED_MC%d_diss.txt'%num)
fig0 = plt.figure(dpi=200)
ax0 = fig0.add_subplot(111)
ax0.grid(True)
_=ax0.set_xlabel("$t/\\tau$",**labfont)
_=ax0.set_ylabel("Norm",**labfont)
_=ax0.plot(Data[:-1,0],Data[:-1,1],ls=":",lw=4)
fig0.savefig("../Pictures/lin_ccqed/%d_norm.png"%num)

tau = 0.1

fig1 = plt.figure(dpi=200)
ax1 = fig1.add_subplot(111)
_=ax1.set_xlabel("$t/\\tau$",**labfont)
_=ax1.set_ylabel("Cavity occupation probabilites",**labfont)
_=ax1.plot(Data2[:-1,0]/tau,Data2[:-1,3],label="MC cav left")
_=ax1.plot(Data2[:-1,0]/tau,Data2[:-1,4],label="MC cav right")
if diss:
    _=ax1.plot(Data3[:-1,0]/tau,Data3[:-1,3],label="MC res left",ls="--")
    _=ax1.plot(Data3[:-1,0]/tau,Data3[:-1,4],label="MC res right",ls="--")
ax1.grid(True)
_=ax1.plot(Data[:-1,0]/tau,Data[:-1,4],ls=":",lw=4,label="MPS left")
_=ax1.plot(Data[:-1,0]/tau,Data[:-1,5],ls=":",lw=4,label="MPS right")
_=ax1.legend()
fig1.savefig("../Pictures/lin_ccqed/%d_cav.png"%num)

fig2 = plt.figure(dpi=200)
ax2 = fig2.add_subplot(111)
_=ax2.set_xlabel("$t/\\tau$",**labfont)
_=ax2.set_ylabel("Atomic excitation probabilites",**labfont)
_=ax2.plot(Data2[:-1,0]/tau,Data2[:-1,1],label="MC cav left")
_=ax2.plot(Data2[:-1,0]/tau,Data2[:-1,2],label="MC cav right")
if diss:
    _=ax2.plot(Data3[:-1,0]/tau,Data3[:-1,1],label="MC res left",ls="--")
    _=ax2.plot(Data3[:-1,0]/tau,Data3[:-1,2],label="MC res right",ls="--")
ax2.grid(True)
_=ax2.plot(Data[:-1,0]/tau,Data[:-1,2],ls=":",lw=4,label="MPS left")
_=ax2.plot(Data[:-1,0]/tau,Data[:-1,3],ls=":",lw=4,label="MPS right")
_=ax2.legend()
fig2.savefig("../Pictures/lin_ccqed/%d_tls.png"%num)

fig3 = plt.figure(dpi=200)
ax3 = fig3.add_subplot(111)
_=ax3.set_xlabel("$t/\\tau$",**labfont)
_=ax3.set_ylabel("$g^{(2)}(t)$",**labfont)
_=ax3.plot(Data[:-1,0]/tau,Data[:-1,6],ls=":",lw=4)
_=ax3.plot(Data[:-1,0]/tau,Data[:-1,7],ls=":",lw=4)
```
